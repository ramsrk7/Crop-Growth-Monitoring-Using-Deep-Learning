```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
image = np.load('image.npy')
label = pd.read_csv('label.csv')
label['1'][306] = 5.0
```

```python
data = pd.DataFrame()
```

```python
data['image'] = image
data['label'] = label
```

```python
data.sample(3)
```

```python
plt.imshow(data.image[1])
```

```python
len(data.image)
```

```python
def remove_noise():
    no_noise = []

    for i in range(len(data.image)):
        blur = cv2.GaussianBlur(data.image[i], (5,5), 0)
        no_noise.append(blur)
    return no_noise
```

```python
data.image = remove_noise()
plt.imshow(data.image[306])
```
